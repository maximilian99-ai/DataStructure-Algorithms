```python
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n - 1) + fibonacci(n - 2)

fibonacci(10)
```

```python
def fibonacci_dp(n):
    cache = [0 for index in range(n + 1)]
    cache[0] = 0
    cache[1] = 1

    for index in range(2, n+1):
        cache[index] = cache[index - 1] + cache[index - 2]
    return cache[n]

fibonacci_dp(10)
```
