```python
import random
data_list = random.sample(range(100), 10)
data_list.sort()

def binary_search(data, search):
    print(data)
    if len(data) == 1:
        if data[0] == search:
            return True
        else:
            return False
    if len(data) == 0:
        return False

    medium = len(data) // 2
    if search == data[medium]:
        return True
    if search > data[medium]:
        return binary_search(data[medium:], search)
    else:
        return binary_search(data[:medium], search)
```

```python
binary_search(data_list, 50)
```
