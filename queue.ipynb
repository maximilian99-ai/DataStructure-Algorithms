```python
queue_list = list()

def enqueue(data):
  queue_list.append(data)

def dequeue(data):
  data = queue_list[0]
  del queue_list[0]
  return data
```

```python
for index in range(10):
  enqueue(index)
```

```python
dequeue(queue_list)
```
