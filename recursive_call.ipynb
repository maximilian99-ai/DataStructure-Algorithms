```python
def factorial(n):
  if n > 1:
    return n * factorial(n -1)
  else:
    return n
```

```python
for n in range(10):
  print(factorial(n))
```

```python
def factorial(n):
  if n <= 1:
    return n
  return_value = n * factorial(n - 1)
  return return_value
```

```python
for n in range(10):
  print(factorial(n))
```

```python
# 다음 함수를 1부터 n까지의 곱을 출력
def multiple(data):
  if data <= 1:
    return data

  return data * multiple(data -1)

for n in range(10):
  print(multiple(n))
```

```python
# 숫자가 들어있는 리스트가 주어졌을 때, 리스트의 합을 리턴하는 함수를 만들기
import random
data = random.sample(range(100), 10)
print(data)
```

```python
def sum_list(data):
  return_value = 0
  for item in data:
    return_value += item
  return return_value

sum_list(data)
```

```python
def sum_list(data):
  if len(data) == 1:
    return data[0]

  return data[0] + sum_list(data[1:])

sum_list(data)
```

```python
# 회문(거꾸로 읽어도 제대로 읽은 것과 같은 단어와 문장)을 판별할 수 있는 함수를 리스트 슬라이싱으로 작성
string = ['D','a','v', 'e']
print(string[-1])
print(string[0])
print(string[1:-1])
print(string[:-1])
```

```python
def recursive(string):
  if len(string) <= 1:
    return True

  if string[0] == string[-1]:
    return recursive(string[1:-1])
  else:
    return False

print(recursive('ruur'))
```
