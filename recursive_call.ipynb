```python
def factorial(n):
    if n > 1:
        return n * factorial(n -1)
    else:
        return n
```

```python
for n in range(10):
    print(factorial(n))
```

```python
def factorial(n):
    if n <= 1:
        return n
    return_value = n * factorial(n - 1)
    return return_value
```

```python
for n in range(10):
    print(factorial(n))
```

```python
def sum_list(data):
    return_value = 0
    for item in data:
        return_value += item
    return return_value

sum_list(data)
```

```python
def sum_list(data):
    if len(data) == 1:
        return data[0]

    return data[0] + sum_list(data[1:])

sum_list(data)
```

```python
def recursive(string):
    if len(string) <= 1:
        return True

    if string[0] == string[-1]:
        return recursive(string[1:-1])
    else:
        return False

print(recursive('ruur'))
```
